```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from pydataset import data
```

### Use the iris database to answer the following quesitons:

```python
iris = sns.load_dataset('iris')
iris.head()
```

#### 1. What does the distribution of petal lengths look like?

```python
sns.distplot(iris.petal_length)
plt.title('Petal Length is a Bimodal Distribution')

plt.show()
```

#### 2. Is there a correlation between petal length and petal width?

```python
sns.relplot(x='petal_width', y='petal_length', data=iris)
plt.title('There is a Postive Correlation Between Petal Length and Width')

plt.show()
```

#### 3. Would it be reasonable to predict species based on sepal width and sepal length?

```python
sns.relplot(x='petal_width', y='petal_length', data=iris, hue='species')
plt.title('There Are Clear Distinctions between Species Petal Characteristics')

plt.show()
```

#### 4. Which features would be best used to predict species?

```python
sns.pairplot(iris, hue='species')

plt.show()
```
